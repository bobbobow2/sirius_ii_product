```python
from sklearn.model_selection import cross_val_score, StratifiedKFold
from sklearn.metrics import precision_score, recall_score, f1_score, roc_auc_score, accuracy_score
from keras.models import Sequential
from keras.layers import Dense

# Функция создания персептрона
def create_perceptron_model():
    model = Sequential()
    model.add(Dense(1, input_dim=X_train.shape[1], activation='sigmoid', kernel_initializer='normal'))
    model.compile(loss='binary_crossentropy', optimizer='adam')
    return model

# Создание персептрона
perceptron_model = create_perceptron_model()

# Организация кросс-валидации
kfold = StratifiedKFold(n_splits=10, shuffle=True, random_state=42)

# Список для сохранения результатов кросс-валидации
results = []

# Проход по каждому фолду и оценка модели
for train_index, val_index in kfold.split(X_train, y_train):
    X_train_fold, X_val_fold = X_train[train_index], X_train[val_index]
    y_train_fold, y_val_fold = y_train[train_index], y_train[val_index]

    # Обучение модели на текущем фолде
    perceptron_model.fit(X_train_fold, y_train_fold, epochs=100, batch_size=10, verbose=0)

    # Предсказание на валидационном наборе данных
    y_val_pred = perceptron_model.predict(X_val_fold)

    # Оценка модели на текущем фолде

# Оценка модели на тестовом наборе данных
y_test_pred = perceptron_model.predict(X_test)

# Вывод метрик на тестовом наборе данных
precision = precision_score(y_test, y_test_pred)
recall = recall_score(y_test, y_test_pred)
f1 = f1_score(y_test, y_test_pred)
roc_auc = roc_auc_score(y_test, y_test_pred)
print("Precision Score:", precision)
print("Recall Score:", recall)
print("F1 Score:", f1)
print("ROC AUC Score:", roc_auc)
```

```python
# The size of encoded and actual representations
encoding_dim = 16
actual_dim = X_train.shape[1]
```

```python
# Input placeholder
input_attrs = Input(shape=(actual_dim,))

# "encoded" is the encoded representation of the input
encoded = Dense(encoding_dim, activation='relu')(input_attrs)

# "decoded" is the lossy reconstruction of the input
decoded = Dense(actual_dim, activation='sigmoid')(encoded)
```

```python
# this model maps an input to its reconstruction
autoencoder = Model(input_attrs, decoded)
```
